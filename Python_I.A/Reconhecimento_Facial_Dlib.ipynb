```python
import dlib
import cv2
import numpy as np
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/drive')
```

```python
img = cv2.imread('/content/drive/MyDrive/M5/Visão Computacional Guia Completo-20220707T140129Z-001/Visão Computacional Guia Completo/Images/people2.jpg')
```

```python
detector_faces_hog = dlib.get_frontal_face_detector()
deteccoes = detector_faces_hog(img, 1) #Escala
```

```python
for face in deteccoes:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(img,(l,t), (r,b), (0,255,255),2)
cv2_imshow(img)
```
